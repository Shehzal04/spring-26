```python
class NQueensSolver:
    """
    Solves the N-Queens problem using backtracking.
    Queens are stored as a 1-D array: board[row] = col.
    """

    def __init__(self, n: int):
        if n <= 0:
            raise ValueError("N must be a positive integer.")
        self.n          = n
        self.board      = [-1] * n   # board[row] = column of queen
        self.solutions  = []         # list of solved boards

    def _is_safe(self, row: int, col: int) -> bool:
        """Return True if (row, col) is not attacked by any previously placed queen."""
        for r in range(row):
            c = self.board[r]
            if c == col or abs(r - row) == abs(c - col):
                return False
        return True

    def _place(self, row: int) -> None:
        """Recursively place queens row by row using backtracking."""
        if row == self.n:                        # all queens placed → solution found
            self.solutions.append(self.board[:])
            return
        for col in range(self.n):
            if self._is_safe(row, col):
                self.board[row] = col
                self._place(row + 1)             # recurse
                self.board[row] = -1             # backtrack


    def solve(self) -> list:
        """Find all solutions and return them."""
        self.solutions.clear()
        self._place(0)
        return self.solutions

    @staticmethod
    def _board_str(solution: list) -> str:
        """Render one solution as a chessboard string."""
        n      = len(solution)
        header = "   " + " ".join(str(c + 1) for c in range(n))
        sep    = "   " + "-" * (2 * n)
        rows   = []
        for r, qcol in enumerate(solution):
            cells = " ".join("Q" if c == qcol else "." for c in range(n))
            rows.append(f"{r + 1} | {cells} |")
        return "\n".join([header, sep] + rows + [sep])

    def display(self, max_show: int = 5) -> None:
        """Print a summary and up to *max_show* board visualisations."""
        n, total = self.n, len(self.solutions)

        print("=" * 50)
        print(f"   {n}-Queens Problem — Results")
        print("=" * 50)

        if not self.solutions:
            print(f"No solution exists for N = {n}.")
            return

        print(f"Total solutions found : {total}")
        show = min(max_show, total)
        print(f"Showing first {show} solution(s):\n")

        for idx in range(show):
            print(f"Solution #{idx + 1}: {self.solutions[idx]}")
            print(self._board_str(self.solutions[idx]))
            print()

        if total > max_show:
            print(f"... and {total - max_show} more solution(s).")

n      = int(input("Enter number of queens (N): "))
solver = NQueensSolver(n)
solver.solve()
solver.display()
```

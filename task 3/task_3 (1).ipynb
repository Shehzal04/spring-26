```python
def get_next_moves(jug1, jug2, max1, max2):
    """
    Generate all valid next states from the current state.
    Returns a list of (new_state, rule_description) tuples.
    """
    moves = []

    if jug1 < max1:
        moves.append(((max1, jug2), f"Fill   Jug1:       ({jug1},{jug2}) → ({max1},{jug2})"))

    if jug2 < max2:
        moves.append(((jug1, max2), f"Fill   Jug2:       ({jug1},{jug2}) → ({jug1},{max2})"))

    if jug1 > 0:
        moves.append(((0, jug2), f"Empty  Jug1:       ({jug1},{jug2}) → (0,{jug2})"))

    if jug2 > 0:
        moves.append(((jug1, 0), f"Empty  Jug2:       ({jug1},{jug2}) → ({jug1},0)"))

    if jug1 > 0 and jug2 < max2:
        pour = min(jug1, max2 - jug2)
        moves.append(((jug1 - pour, jug2 + pour),
                      f"Pour   Jug1→Jug2:  ({jug1},{jug2}) → ({jug1-pour},{jug2+pour})"))

    if jug2 > 0 and jug1 < max1:
        pour = min(jug2, max1 - jug1)
        moves.append(((jug1 + pour, jug2 - pour),
                      f"Pour   Jug2→Jug1:  ({jug1},{jug2}) → ({jug1+pour},{jug2-pour})"))

    return moves


def water_jug_dfs(max1=4, max2=3, target=2):
    """
    Solve the Water Jug Problem using Depth-First Search (DFS).

    Args:
        max1   : capacity of jug 1 (default 4L)
        max2   : capacity of jug 2 (default 3L)
        target : desired amount of water (default 2L)

    Returns:
        True if solution found, False otherwise.
    """
    stack   = [((0, 0), [])]
    visited = set()         

    print("=" * 50)
    print("   Water Jug Problem — DFS Solution")
    print("=" * 50)
    print(f"  Jug1 capacity : {max1}L")
    print(f"  Jug2 capacity : {max2}L")
    print(f"  Target        : {target}L")
    print("=" * 50)
    print(f"{'Step':<6} {'Jug1':>6} {'Jug2':>6}")
    print("-" * 22)

    step = 1

    while stack:
        (jug1, jug2), rules = stack.pop()

        if (jug1, jug2) in visited:
            continue
        visited.add((jug1, jug2))        

        print(f"{step:<6} {jug1:>6} {jug2:>6}")
        step += 1

        # Goal check
        if jug1 == target or jug2 == target:
            print("-" * 22)
            print("\nSOLUTION FOUND!\n")
            print(f"{'#':<4} {'Rule':<40} {'State After'}")
            print("-" * 65)
            for i, rule in enumerate(rules, 1):
                state_after = rule.split("→")[-1].strip()
                print(f"{i:<4} {rule.split('→')[0].strip():<40} → {state_after}")
            print("-" * 65)
            print(f"\nTotal steps to solution : {len(rules)}")
            print(f"Total states explored   : {len(visited)}")
            return True

        next_moves = get_next_moves(jug1, jug2, max1, max2)
        for state, rule in reversed(next_moves):
            if state not in visited:
                stack.append((state, rules + [rule]))

    print("\n No solution exists.")
    return False

water_jug_dfs(max1=4, max2=3, target=2)
```
